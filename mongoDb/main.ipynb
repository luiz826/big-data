```python
import csv
import os

from pymongo import MongoClient
from dotenv import load_dotenv
from datetime import datetime

load_dotenv()
```

```python
mongo_key = os.getenv("MONGO_KEY")
```

```python
client = MongoClient(mongo_key)

db = client['loteria_0']
megasena = db['megasena']

megasena.count_documents({})
```

```python
csv_file_path = "./data/megasena.csv"

if megasena.count_documents({}) == 0:
    with open(csv_file_path, mode='r') as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            megasena.insert_one(row)
```

```python
for doc in megasena.find({"Ganhadores_Quina": {"$type": "string"}}):
    try:
        ganhadores_quina_int = int(doc["Ganhadores_Quina"])
        megasena.update_one(
            {"_id": doc["_id"]},
            {"$set": {"Ganhadores_Quina": ganhadores_quina_int}}
        )
    except ValueError:
        print(f"Falha ao converter Ganhadores_Quina para o concurso {doc['Concurso']}")
```

```python
for doc in megasena.find({"Data Sorteio": {"$type": "string"}}):
    try:
        data_sorteio_date = datetime.strptime(doc["Data Sorteio"], "%d/%m/%Y")
        megasena.update_one(
            {"_id": doc["_id"]},
            {"$set": {"Data Sorteio": data_sorteio_date}}
        )
    except ValueError:
        pass
```

```python
for doc in megasena.find({"Arrecadacao_Total": {"$type": "string"}}):
    try:
        new_val = float(doc["Arrecadacao_Total"])
        megasena.update_one(
            {"_id": doc["_id"]},
            {"$set": {"Arrecadacao_Total": new_val}}
        )
    except ValueError:
        pass
```

```python
for doc in megasena.find({"Ganhadores_Sena": {"$type": "string"}}):
    try:
        new_val = int(doc["Ganhadores_Sena"])
        megasena.update_one(
            {"_id": doc["_id"]},
            {"$set": {"Ganhadores_Sena": new_val}}
        )
    except ValueError:
        pass
```

```python
query_result = megasena.find({"Concurso": concurso_id})
print_result(query_result)
```

# 3. Listar Todos os Concursos com Acumulado

```python
query_result = megasena.find({"Acumulado": "SIM"})
print_result(query_result)
```

# 4. Atualizar a Arrecadação Total de um Concurso

```python

numero_concurso = concurso_id 
nova_arrecadacao = 60000000.00

result = megasena.update_one(
    {"Concurso": concurso_id},
    {"$set": {"Arrecadacao_Total": nova_arrecadacao}}
)

if result.modified_count > 0:
    print(f"A arrecadação total do concurso {numero_concurso} foi atualizada com sucesso.")
else:
    print(f"Nenhuma atualização foi feita para o concurso {numero_concurso}.")
```

# 5. Acrescentar uma Cidade e UF aos Ganhadores da Sena

```python
cidade = "Mamanguape"
uf = "PB"

result = megasena.update_one(
    {
        "Concurso": "2", 
        "Ganhadores_Sena": {"$ne": "0"}  # Garante que o concurso teve ganhadores da Sena (diferente de "0")
    },
    {
        "$set": {
            "Cidade": cidade,
            "UF": uf
        }
    }
)
```

```python
query_result = megasena.find({"Concurso": "2", "Ganhadores_Sena": {"$ne": "0"}})
print_result(query_result)
```

# 7. Excluir Todos os Concursos Sem Ganhadores da Sena

```python
result = megasena.delete_many({"Ganhadores_Sena": "0"})

if result.deleted_count > 0:
    print(f"Foram excluídos {result.deleted_count} concursos sem ganhadores da Sena.")
else:
    print("Nenhum concurso sem ganhadores da Sena foi encontrado.")
```

# 8. Buscar Concursos com Número de Ganhadores da Quina Acima de um Certo Limite

```python
query_result = megasena.find({"Ganhadores_Quina": {"$gt": 50}})
print_result(query_result)
```

# 9. Buscar Concursos em um Intervalo de Datas

```python
data_inicio = datetime(2000, 1, 1)  # Data de início
data_fim = datetime(2010, 1, 1)    # Data de fim

query_result = megasena.find({
    "Data Sorteio": {
        "$gte": data_inicio,  # Maior ou igual à data de início
        "$lte": data_fim      # Menor ou igual à data de fim
    }
})

print_result(query_result)
```

# 10. Buscar Concursos com Arrecadação Total Acima de um Valor e Menos de X Ganhadores da Sena

```python
valor_minimo_arrecadacao = 1000000.00 
max_ganhadores_sena = 2  

query_result = megasena.find({
    "Arrecadacao_Total": {"$gt": valor_minimo_arrecadacao},
    "Ganhadores_Sena": {"$lt": max_ganhadores_sena}
})

print_result(query_result)
```

# 11. Buscar Concursos onde o Valor Acumulado está Entre Dois Valores

```python
min_acum = 500000.00 
max_acum = 1000000.00 

query_result = megasena.find({
    "Valor_Acumulado": {"$gt": min_acum},
    "Valor_Acumulado": {"$lt": max_acum}
})

print_result(query_result)
```
